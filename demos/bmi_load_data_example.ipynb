```python
import numpy as np
import pickle as pkl
from google.colab import drive
import matplotlib.pyplot as plt
```

```python
# Mount your Google Drive into the local virtual machine file system (will give a warning if already mounted)
# If asked, follow the directions to authenticate with your Google Drive
drive.mount('/content/drive')
```

```python
# Load a brain-machine interface dataset
fname = '/content/drive/MyDrive/MLP_2021/datasets/bmi_dataset.pkl'
with open(fname, 'rb') as fp:
  dat = pkl.load(fp)
```

```python
# Dat is a dictionary
dat.keys()
```

```python
# 20 non-overlapping data sets: joint angle
len(dat['theta'])
```

```python
# Shape of the 0th data set
dat['theta'][0].shape
```

```python
# Extract joint angles and timestamps from the 0th data set
theta = dat['theta'][0]
tm = dat['time'][0]

# We will focus on just the first 300 samples in the timeseries
inds = range(300)
```

```python
# Plot both shoulder (0) and elbow (1) as a function of time
plt.plot(tm[inds,:], theta[inds,:])
plt.legend(['shoulder', 'elbow'])
plt.xlabel('Time (s)')
plt.ylabel('theta (rad)')
```
